# Normalise the Dataset

We now need to figure out the line of this dataset.
We have to lern the value of theta where theta is a vector
theta = [01, 02]

Loss Function:E = Sum((Y'i - Yi)** 2)

dE/d01 = Sum(2*(Y'i - Yi))

dE/d02 = Sum(2*(Y'i - Yi)) * d(Y'i)/d01
                                       where d(Y'i)/d01 = X

Thus, dE/d02 = Sum(2*(Y'i - Yi)) * X

Hence,
    01 = 01 - n * dE/d01
    02 = 02 - n * dE/d02

We can visualize it as a 3D graph with Error, 01 and 02 as axis in which a 3D bowl will be created and we will slide down to the minima of the bowl 

```python
def gradient(X,Y,theta):
    
    m = X.shape[0] #Gives the number of training points
    grad = np.zeros((2,))
    
    for i in range(X.shape[0]):
        Yi = theta[1]*X[i] + theta[0]
        grad[0] += 2*(Yi - Y[i])
        grad[1] += 2*(Yi - Y[i])*X[i]
    
    return grad
```

```python
def total_error(X,Y,theta):
    
    m = X.shape[0]
    e = 0
    for i in range(m):
        Yi = theta[1]*X[i] + theta[0]
        e += (Yi - Y[i])**2
    
    return e
```

```python
def gradientdescent (X, Y, learning_rate = 0.001, max_steps = 50) :
    
    theta = np.zeros((2,))
    error_list = []
    
    for i in range(max_steps):
        #Update theta
        grad = gradient(X, Y, theta)
        theta[0] = theta[0] - learning_rate * grad[0]
        theta[1] = theta[1] - learning_rate * grad[1]
        error = total_error(X,Y,theta)
    
        error_list.append(error)
        
    return theta, error_list
```

```python
theta,error_list = gradientdescent (X, Y)
```
